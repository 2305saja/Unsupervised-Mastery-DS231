**رسمة الداتا قبل PCA**

```python
import numpy as np
Card_meaned = Card_Data - np.mean(Card_Data, axis=0)
X_std = Card_meaned / np.std(Card_meaned, axis=0)

cov_matrix = np.cov(X_std, rowvar=False)

eigenvalues, eigenvectors = np.linalg.eigh(cov_matrix)

sorted_index = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_index]

explained_variance_ratio = sorted_eigenvalues / np.sum(sorted_eigenvalues)

cumulative_explained_variance = np.cumsum(explained_variance_ratio)

for i, (var, cum_var) in enumerate(zip(explained_variance_ratio, cumulative_explained_variance), start=1):
    print(f"PC{i}: Explained Variance Ratio = {var:.4f}, Cumulative = {cum_var:.4f}")
```

**Variance Ratio حساب **

```python
threshold = 0.90
num_components = np.argmax(cumulative_explained_variance >= threshold) + 1

print(f"Number of components to preserve {threshold*100}% variance: {num_components}")
```

```python
threshold = 0.95
num_components = np.argmax(cumulative_explained_variance >= threshold) + 1

print(f"Number of components to preserve {threshold*100}% variance: {num_components}")
```

##2. Clustering Analysis##

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

k_values = [4, 5, 6]
Inertia_Value = []
silhouette_scores = []

for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(Card_Data_pca)

    Inertia_Value.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(Card_Data_pca, labels))

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

ax1.plot(k_values, Inertia_Value, marker='o', color='skyblue')
ax1.set_title('Elbow Method: Inertia vs. Number of Clusters (k)')
ax1.set_xlabel('Number of clusters (k)')
ax1.set_ylabel('Inertia')
ax1.set_xticks(k_values)
ax1.grid(True)

ax2.plot(k_values, silhouette_scores, marker='o', color='orange')
ax2.set_title('Silhouette Score vs. Number of Clusters (k)')
ax2.set_xlabel('Number of clusters (k)')
ax2.set_ylabel('Silhouette Score')
ax2.grid(True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import silhouette_samples
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans

cluster_numbers = [4, 5, 6]
for k in cluster_numbers:


    fig, ax = plt.subplots(figsize=(10, 8))

    data_to_cluster = Card_Data_pca


    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(data_to_cluster)

    scores = silhouette_samples(data_to_cluster, labels)
    avg_score = np.mean(scores)


    ax.set_title(f"Silhouette Analysis (k={k})", pad=20)
    ax.set_xlabel("Silhouette Score")
    ax.axvline(avg_score, color='pink', linestyle='--')
    ax.set_xlim([-0.1, 1])
    ax.set_ylim([0, len(Card_Data) + (k + 1) * 10])
    ax.grid(True, linestyle='--', alpha=0.5)

    y_position = 10

    for cluster in range(k):
        cluster_scores = sorted(scores[labels == cluster])
        y_end = y_position + len(cluster_scores)

        ax.fill_betweenx(
            np.arange(y_position, y_end),
            0, cluster_scores,
            color=plt.cm.tab10(cluster / k),
            alpha=0.9
        )
        ax.text(-0.08, y_position + len(cluster_scores) / 2, str(cluster))
        y_position = y_end + 10

    plt.tight_layout()
    plt.show()
```

```python
from sklearn.cluster import DBSCAN

eps_values = [0.3, 0.5, 0.7, 0.9, 1.1]
min_samples_values = [3, 5, 7, 10]

results = []

for eps in eps_values:
    for min_samples in min_samples_values:
        db = DBSCAN(eps=eps, min_samples=min_samples)
        labels = db.fit_predict(Card_Data_pca)

        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        n_noise = list(labels).count(-1)

        if n_clusters > 1:
            score = silhouette_score(Card_Data_pca, labels)
        else:
            score = -1

        results.append({
            'eps': eps,
            'min_samples': min_samples,
            'clusters': n_clusters,
            'noise': n_noise,
            'silhouette': round(score, 3)
        })

df_results = pd.DataFrame(results)
df_results = df_results.sort_values(by='silhouette', ascending=False)
```

##3. Anomaly Detection##

```python
from sklearn.ensemble import IsolationForest
from sklearn.svm import OneClassSVM
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score, f1_score

Card_WOC = Card_Data.drop('Class', axis=1)
Class_Card = Card_Data['Class']

contamination = Class_Card.sum() / len(Class_Card)

forest = IsolationForest(contamination=contamination, random_state=42)
forest.fit(Card_Data[Bef_Scaling])
Class_iso = forest.predict(Card_Data[Bef_Scaling])

Class_iso= [1 if Card_WOC== -1 else 0 for Card_WOC in Class_iso]

print("=== Isolation Forest Performance ===")
print(classification_report(Class_Card, Class_iso))
print("Confusion Matrix:\n", confusion_matrix(Class_Card, Class_iso))

oc_svm = OneClassSVM(nu=contamination, kernel='rbf', gamma='scale')
oc_svm.fit(Card_Data[Bef_Scaling])
Class_svm = oc_svm.predict(Card_Data[Bef_Scaling])
Class_svm = [1 if x == -1 else 0 for x in Class_svm]

print("=== One-Class SVM Performance ===")
print(classification_report(Class_Card, Class_svm))
print("Confusion Matrix:\n", confusion_matrix(Class_Card, Class_svm))

precision_iso = precision_score(Class_Card, Class_iso)
recall_iso = recall_score(Class_Card, Class_iso)
f1_iso = f1_score(Class_Card, Class_iso)

precision_svm = precision_score(Class_Card, Class_svm)
recall_svm = recall_score(Class_Card, Class_svm)
f1_svm = f1_score(Class_Card, Class_svm)

print(f"Isolation Forest - Precision: {precision_iso:.4f}, Recall: {recall_iso:.4f}, F1-Score: {f1_iso:.4f}")
print(f"One-Class SVM    - Precision: {precision_svm:.4f}, Recall: {recall_svm:.4f}, F1-Score: {f1_svm:.4f}")
```
